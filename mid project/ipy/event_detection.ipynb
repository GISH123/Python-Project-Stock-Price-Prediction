```python
def stock_return(price_now, price_past):
    return (price_now-price_past) / price_past
```

```python
import numpy as np
for df_i in df_list:
    price_list = list(df_i[u'收盤價(元)'])
    return_list = [0]
    for i in range(len(price_list)):
        price_now = price_list[i]
        if i !=0:
            a_return = stock_return(price_now, price_last)
            return_list.append(a_return)
        price_last = price_now
    df_i['return'] = np.array(return_list)
```

```python
mean_list = {}
std_list = {}

for df_i in df_list:
    a_id = df_i.index[0]
    std_list[a_id] = df_i['return'].std()
    mean_list[a_id] = df_i['return'].mean()
```

```python
event_dict = {}
for df_i in df_list:
    
    a_id = df_i.index[0]
    event_dict[a_id] = []
    
    
    date_list = list(df_i[u'年月日'])
    return_list = list(df_i['return'])
    upper_bar = mean_list[a_id] + std_list[a_id]
    lower_bar = mean_list[a_id]- std_list[a_id]
    
    
    for index in range(1, len(return_list)):
        a_return = return_list[index]
        if a_return>upper_bar :
            event_dict[a_id].append(  (date_list[index].date() , 'up'  ))
        elif a_return< lower_bar:
            event_dict[a_id].append(( date_list[index].date() , 'down'  ))
```

```python
len_dict = dict()
len_dict['news'] = {'all' :232514 , 'U': 7707, 'D': 1705}
```

```python
dict_of_dict = dict() #  dict_of_dict[news][up, down, flat] = dict() of chi2 value
import csv
def get_chi2_dictionary():
    for item in ['news']:
        dict_all = dict()
        dict_of_dict[item]  = dict()
        
        coding_style = 'big5' if item == 'news' else 'utf-8'
        # all dictionary
        with open( r'.\data' + '\\' + item + '_all_grams' + '.csv' , 'r', encoding = coding_style) as f:
            csv_reader = csv.DictReader(f)
            for a_row in csv_reader:
                df_value = int(a_row['DF'])
                word = a_row[u'單詞']
                dict_all[word] = df_value
                
                
       
        for a_class in ['U', 'D']:
            dict_of_dict[item][a_class] = dict()
            
            # up and down and flat
            
            with open( r'.\data' + '\\' + item + '_' + a_class +'_grams' + '.csv' , 'r', encoding = coding_style) as f:
                csv_reader = csv.DictReader(f)
                for a_row in csv_reader:
                    df_value = int(a_row['DF'])
                    word = a_row[u'單詞']
                    if word not in  dict_all:
                        continue
                    else:
                        expected_df_value = dict_all[word] * len_dict[item][a_class] / len_dict[item]['all']
                        sign = 1 if df_value >= expected_df_value else -1
                        dict_of_dict[item][a_class][word] = sign * (  (df_value - expected_df_value) **2 ) / expected_df_value
get_chi2_dictionary()
```

```python
def drop_subwords(list1): 

    output_list=  [1]
    dropped_list = []
    dict1 = {}
    while len(output_list) >0:
        output_list=  []
        dropped_list = []
        dict1 = {}
        for i in range(len(list1)):

            word = list1[i]
            length = len(word)

            sub_word1 =  word[0:length-1] 
            sub_word2 = word[1:length] 	 
            
            if word == '──────' or word in id_list:
                dropped_list.append(word)
            elif word == '▇▇▇▇▇▇':
                dropped_list.append(word)
            
            
            
            if length <5 or sub_word1 == sub_word2 :
                continue



            if sub_word1 not in dict1:
                dict1[sub_word1] = (i, 0)
            elif dict1[sub_word1][1] == 1:

                merged_word = list1[ dict1[sub_word1][0]	][0] + word
                output_list.append(			merged_word				)
                dropped_list.append(   list1[i] 			)
                dropped_list.append(   list1[ dict1[sub_word1][0]	]	)

            if sub_word2 not in dict1:
                dict1[sub_word2] = (i, 1)
            elif dict1[sub_word2][1] == 0:

                merged_word = word[0] + list1[ dict1[sub_word2][0]	]
                output_list.append(			merged_word				)
                dropped_list.append(   list1[i] 			)
                dropped_list.append(   list1[ dict1[sub_word2][0]	]	)
        #print(dropped_list)
        #print(output_list)
        for i in set(dropped_list):
            list1.remove(i)
        for i in set(output_list):
            list1.append(i)

    return list1
```

```python
import csv 
from datetime import datetime
from datetime import timedelta

for item in new_id_list:
    
    target = dict()

    a_id = item.split()[0]
    mean, std = (mean_list[item], std_list[item])

    file_price = open('stock_price_%s.csv' %a_id, 'r')
    reader = csv.DictReader(file_price)
    price_date_list_1 = []
    price_date_list_2 = []

    for a_row in reader:
        date = datetime.strptime( a_row['date'], "%Y-%m-%d").date()
        returns = float(a_row['return'])


        price_date_list_1.append(date)
        price_date_list_2.append(date + timedelta(days = -1))

        if returns <  mean - std  : 
            target[date] = 0
        elif returns > mean  +  std :
            target[date] = 2
        else:
            target[date] = 1




    news_content_plus_title = dict()
    news_number_of_document = dict()
    '''
    news_content_plus_title[date] = '文章字串'
    '''

    time_delta = timedelta(hours = -14)


    with open( r'C:\Users\User\Documents\GitHub\Big-Data-and-Business-Analytics\csv_output_%s.csv' %a_id  , 'r', encoding = 'big5') as f:
        #use market date  to create a dictionary
        for date in price_date_list_1:
            news_content_plus_title[date] = {'U':'', 'flat':'', 'D':''}
            news_number_of_document[date] = {'U':0, 'flat':0, 'D':0}


        csv_reader = csv.DictReader(f)
        for a_row in csv_reader:

            time = datetime.strptime( a_row['post_time'], "%Y/%m/%d %H:%M")
            label = label_dict[int(float(a_row['label']))]
            title = a_row['title']
            content = a_row['content']
            title_plus_content = title + '＜' + content
            
            adj_time = time+time_delta
            adj_date = (adj_time + timedelta(days = 1)) #新聞的後一天



            if (adj_date.date() in price_date_list_2) : #後兩天股價
                adj_date = (adj_time + timedelta(days = 2)).date()
                news_content_plus_title[adj_date][label] +=  '＜' + title_plus_content

                news_number_of_document[adj_date][label]  += 1      

            elif (adj_date.date() in price_date_list_1 and int(adj_date.hour) in  range(0, 19)) : #隔天股價
                adj_date = (adj_time + timedelta(days = 1)).date()


                news_content_plus_title[adj_date][label] +=  '＜' + title_plus_content

                news_number_of_document[adj_date][label]  += 1
                
                

    features_dict = {}
    for date in price_date_list_1:
        features_dict[date] = [] 



    article = news_content_plus_title
    article_number = news_number_of_document

    feature_list = []
    for a_class in ['U', 'D']:
        keyword_list = df['news'+ '_' + a_class].dropna().tolist()


        feature_list += keyword_list

        for date in price_date_list_1:
            for keyword in keyword_list:

                try:
                    feature_i = article[date][a_class].count(keyword) / article_number[date][a_class]
                except:
                    feature_i = 0

                features_dict[date].append(    feature_i )         

    df_ready = pd.DataFrame.from_dict(features_dict, orient = 'index', columns  = feature_list)




    y = pd.DataFrame.from_dict(target, orient = 'index', columns  = ['target'])
    df_ready['target'] = y
    data_df_dict[a_id] = df_ready
```

```python
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score 
from sklearn.metrics import classification_report
from sklearn.neighbors import KNeighborsClassifier
```

```python
def predict( stock_id):
    res = data_df_dict[stock_id]
    y = res.iloc[:,:]['target']
    X = res.iloc[:,0:-1]
    x_train,x_test,y_train,y_test = tt_split(X,y,test_size=0.2,random_state=0)

    clf_list = []
    rf = RandomForestClassifier( criterion =  'entropy' ,  random_state = 42)
    logistic = LogisticRegression()
    knn = KNeighborsClassifier()
    clf_list += [rf,logistic, knn ]
    
    y_true = list(y_test)
    target_names = ['class 0', 'class 1', 'class 2']
    predictlist = []
    for clf in clf_list:
        clf.fit(x_train, y_train)
        y_pred = list( clf.predict(x_test) )
        predictlist.append(y_pred)
        print(classification_report(y_true, y_pred, target_names=target_names, digits = 4))    
```
