```python
len_dict = dict()
len_dict['news'] = {'all' :232514 , 'up': 70719, 'down': 38381, 'flat': 123414}
len_dict['bbs'] = {'all' :57930 , 'up': 13339, 'down': 10010, 'flat': 34581}
len_dict['forum'] = {'all' :25562 , 'up': 453, 'down': 381, 'flat': 24728}
```

```python
dict_of_dict = dict() #  dict_of_dict[news, forum, bbs][up, down, flat] = dict() of chi2 value
import csv
def get_chi2_dictionary():
    for item in ['news', 'bbs', 'forum']:
        dict_all = dict()
        dict_of_dict[item]  = dict()
        
        coding_style = 'big5' if item == 'news' else 'utf-8'
        # all dictionary
        with open( r'.\data' + '\\' + item + '_all_grams' + '.csv' , 'r', encoding = coding_style) as f:
            csv_reader = csv.DictReader(f)
            for a_row in csv_reader:
                df_value = int(a_row['DF'])
                word = a_row[u'單詞']
                dict_all[word] = df_value
                
                
       
        for a_class in ['up', 'flat', 'down']:
            dict_of_dict[item][a_class] = dict()
            
            # up and down and flat
            
            with open( r'.\data' + '\\' + item + '_' + a_class +'_grams' + '.csv' , 'r', encoding = coding_style) as f:
                csv_reader = csv.DictReader(f)
                for a_row in csv_reader:
                    df_value = int(a_row['DF'])
                    word = a_row[u'單詞']
                    if word not in  dict_all:
                        continue
                    else:
                        expected_df_value = dict_all[word] * len_dict[item][a_class] / len_dict[item]['all']
                        sign = 1 if df_value >= expected_df_value else -1
                        dict_of_dict[item][a_class][word] = sign * (  (df_value - expected_df_value) **2 ) / expected_df_value
```

```python
def drop_subwords(list1): 

    output_list=  [1]
    dropped_list = []
    dict1 = {}
    while len(output_list) >0:
        output_list=  []
        dropped_list = []
        dict1 = {}
        for i in range(len(list1)):

            word = list1[i]
            length = len(word)

            sub_word1 =  word[0:length-1] 
            sub_word2 = word[1:length] 	 
            
            if word == '──────' :
                dropped_list.append(word)
            elif word == '▇▇▇▇▇▇':
                dropped_list.append(word)
            
            
            
            if length <5 or sub_word1 == sub_word2 :
                continue



            if sub_word1 not in dict1:
                dict1[sub_word1] = (i, 0)
            elif dict1[sub_word1][1] == 1:

                merged_word = list1[ dict1[sub_word1][0]	][0] + word
                output_list.append(			merged_word				)
                dropped_list.append(   list1[i] 			)
                dropped_list.append(   list1[ dict1[sub_word1][0]	]	)

            if sub_word2 not in dict1:
                dict1[sub_word2] = (i, 1)
            elif dict1[sub_word2][1] == 0:

                merged_word = word[0] + list1[ dict1[sub_word2][0]	]
                output_list.append(			merged_word				)
                dropped_list.append(   list1[i] 			)
                dropped_list.append(   list1[ dict1[sub_word2][0]	]	)
        #print(dropped_list)
        #print(output_list)
        for i in set(dropped_list):
            list1.remove(i)
        for i in set(output_list):
            list1.append(i)

    return list1
```

```python
import pandas as pd
keyword_dict = {}
for item in ['news', 'bbs', 'forum']:           
    for a_class in ['up', 'flat', 'down']:
        count = 100
        index = pd.DataFrame.from_dict(  dict_of_dict[item][a_class], orient='index', columns=[item + '_' +a_class]  ).sort_values(by=item + '_' +a_class, ascending=False).index
        keyword_list = list(index[0:100])
        keyword_list = drop_subwords(keyword_list)
        length = len(keyword_list)
        count_next = 100 + (100-length)
        while length <100:
            
            keyword_list += list( index[ count: count_next] ) 
            keyword_list = drop_subwords(keyword_list)
            length = len(keyword_list)
            count, count_next =( count_next, count_next + 100-length )
            print(length, count, count_next)
            
        keyword_dict[item + '_' +a_class] =  keyword_list
```

```python
target = dict()
ma_dict = dict()
import csv 
from datetime import datetime
from datetime import timedelta

file_market = open('market_ready.csv', 'r')
reader_market = csv.DictReader(file_market)
market_date_list = []


for a_row in reader_market:
    date = datetime.strptime( a_row['date'], "%Y-%m-%d").date()
    two_day_returns = float(a_row[u'2日報酬'])
    price = float(a_row['price'])
    ma = float(a_row['5MA'])
    market_date_list.append(date)
    ma_dict[date] = (price-ma)/ma
    if two_day_returns <= 0 : 
        target[date] = 0
    elif two_day_returns > 0 :
        target[date] = 1
```

```python
label_dict = {2:'up', 1:'flat', 0:'down'}
```

```python
import csv 
from datetime import datetime
from datetime import timedelta

bbs_content_plus_title = dict()
bbs_number_of_document = dict()
'''
bbs_content_plus_title[date] = '文章字串'
'''

time_delta = timedelta(hours = -13.5)
item = 'bbs'

with open( r'.\data' + '\\' + item + '_all' + '.csv' , 'r', encoding = 'utf-8') as f:
    #use market date  to create a dictionary
    for date in market_date_list:
        bbs_content_plus_title[date] = {'up':'', 'flat':'', 'down':''}
        bbs_number_of_document[date] = {'up':0, 'flat':0, 'down':0}


    csv_reader = csv.DictReader(f)
    for a_row in csv_reader:
        time = datetime.strptime( a_row['post_time'], "%Y/%m/%d %H:%M:%S") 
        adj_time = time+time_delta
        adj_date = (adj_time + timedelta(days = 2)).date()


        if  adj_date in market_date_list: 
            title = a_row['title']
            content = a_row['content']
            label = label_dict[int(float(a_row['label']))]
            title_plus_content = title + '＜' + content

            
            bbs_content_plus_title[adj_date][label] +=  '＜' + title_plus_content

            bbs_number_of_document[adj_date][label]  += 1
```
