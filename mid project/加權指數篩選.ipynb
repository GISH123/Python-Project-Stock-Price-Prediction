# 計算加權指數的1日至5日報酬率，計算5日MA

```python
#函數定義
def moving_average(lst): 
    return sum(lst) / len(lst) 
def stock_return(price_now, price_past):
    return (price_now-price_past) / price_past
```

```python
import csv 
from datetime import datetime
file = open('market_price.csv', 'r')
reader = csv.DictReader(file)
five_day_price = []

data_list = []

for a_row in reader:
        tem_dict = dict()
        
        date = datetime.strptime( a_row[u'年月日'], "%Y-%m-%d")
        price = float(a_row[u'收盤價(元)'] )
        five_day_price.append( price )
        
        

        
        
        
        if len(five_day_price) == 6:
            #計算1日到過去5日報酬
            returns = []
            for i in range(5):
                returns.append(stock_return(price, five_day_price[i] )  )
            
                                  
            
        
            #計算MA
            five_day_price = five_day_price[1:] #保留5個
            five_MA = moving_average(five_day_price)


            #存成dictionary
            tem_dict['name'] = 'Y9999 加權指數'
            tem_dict['date'] = date
            tem_dict['price'] = price
            tem_dict['5MA'] = five_MA
            tem_dict['1日報酬'] = returns[4]
            tem_dict['2日報酬'] = returns[3]
            tem_dict['3日報酬'] = returns[2]
            tem_dict['4日報酬'] = returns[1]
            tem_dict['5日報酬'] = returns[0]
        

            data_list.append(tem_dict)
            
        
        
```

## 將list轉成dataframe方便輸出

```python
import pandas as pd
df = pd.DataFrame(data_list)[['name', 'date', 'price', '5MA', u'1日報酬',  u'2日報酬',  u'3日報酬',  u'4日報酬',  u'5日報酬',]]
```

## 因為樣本數為732，我想要選出報酬分布前10%及後10%當作漲及跌的判斷標準

```python
import  matplotlib.pyplot as plt
def draw_r_distribution(string):
    string = str(string)
    x1 = df[u'%s日報酬' %string].sort_values(ascending=True)
    #報酬倒數73個樣本的報酬率
    down_r = float(x1.iloc[72])
    #報酬第73個樣本的報酬率 732-73 = 659
    up_r = float(x1.iloc[659])

    #界線
    x0 = down_r
    y0 = 100
    plt.plot([x0, x0,], [0, y0,], 'k--', linewidth=2.5)
    plt.text(-0.05, 100 , s = 'The 10 percent %s day return  is %f' %( string, float(x1.iloc[72]) ),
             fontdict={'size': 16, 'color': 'r'} )


    x0 = up_r
    y0 = 100
    plt.plot([x0, x0,], [0, y0,], 'k--', linewidth=2.5)
    plt.text(-0.05, 120 , s = 'The 90 percent %s day return  is %f' %(string, float(x1.iloc[659]) ),
             fontdict={'size': 16, 'color': 'r'} )


    n, bins, patches = plt.hist(x=x1, bins='auto', color='#0504aa',  alpha=0.7, rwidth=0.85)
```

```python
draw_r_distribution(1)
```
